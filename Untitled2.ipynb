#L√†m s·∫°ch vƒÉn b·∫£n

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)  # x√≥a link
    text = re.sub(r'[^a-z0-9√°√†·∫£√£·∫°ƒÉ·∫Ø·∫±·∫≥·∫µ·∫∑√¢·∫•·∫ß·∫©·∫´·∫≠√©√®·∫ª·∫Ω·∫π√™·∫ø·ªÅ·ªÉ·ªÖ·ªá√≠√¨·ªâƒ©·ªã√≥√≤·ªè√µ·ªç√¥·ªë·ªì·ªï·ªó·ªô∆°·ªõ·ªù·ªü·ª°·ª£√∫√π·ªß≈©·ª•∆∞·ª©·ª´·ª≠·ªØ·ª±√Ω·ª≥·ª∑·ªπ·ªµƒë ]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['clean_review'] = df['content'].apply(clean_text)
```

#G√°n nh√£n c·∫£m x√∫c (Sentiment)


```python
def map_sentiment(score):
    if score <= 2:
        return 'negative'
    elif score == 3:
        return 'neutral'
    else:
        return 'positive'

df['sentiment'] = df['score'].apply(map_sentiment)
```

# Chia train/test

#C√†i ƒë·∫∑t c√°c th∆∞ vi·ªán c·∫ßn thi·∫øt

```python
from transformers import DistilBertTokenizerFast, DistilBertForSequenceClassification

# T·∫£i tokenizer v√† m√¥ h√¨nh DistilBERT
tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')
model = DistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=3)  # 3: negative, neutral, positive
model = model.to('cuda')  # Chuy·ªÉn m√¥ h√¨nh sang GPU
```

# Chu·∫©n b·ªã d·ªØ li·ªáu

```python
from torch.utils.data import Dataset, DataLoader
import torch

class SentimentDataset(Dataset):
    def __init__(self, reviews, labels, tokenizer, max_len):
        self.reviews = reviews
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.reviews)

    def __getitem__(self, item):
        review = self.reviews[item]
        label = self.labels[item]
        encoding = self.tokenizer.encode_plus(
            review,
            add_special_tokens=True,
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt',
        )
        return {
            'review_text': review,
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }

# Chu·∫©n b·ªã d·ªØ li·ªáu train v√† test
train_dataset = SentimentDataset(
    reviews=train_df['clean_review'].to_numpy(),
    labels=train_df['sentiment'].map({'positive': 2, 'neutral': 1, 'negative': 0}).to_numpy(),
    tokenizer=tokenizer,
    max_len=128  # S·ª≠ d·ª•ng ƒë·ªô d√†i t·ªëi ƒëa 128 t·ª´ ƒë·ªÉ t·ªëi ∆∞u b·ªô nh·ªõ GPU
)

test_dataset = SentimentDataset(
    reviews=test_df['clean_review'].to_numpy(),
    labels=test_df['sentiment'].map({'positive': 2, 'neutral': 1, 'negative': 0}).to_numpy(),
    tokenizer=tokenizer,
    max_len=128  # S·ª≠ d·ª•ng ƒë·ªô d√†i t·ªëi ƒëa 128 t·ª´ ƒë·ªÉ t·ªëi ∆∞u b·ªô nh·ªõ GPU
)
```

#Hu·∫•n luy·ªán m√¥ h√¨nh BERT

```python
train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=128)
```

```python
from torch.optim import AdamW
from tqdm import tqdm

optimizer = AdamW(model.parameters(), lr=2e-5)  # Learning rate nh·ªè ƒë·ªÉ tr√°nh overfitting

# Hu·∫•n luy·ªán m√¥ h√¨nh
def train_epoch(model, data_loader, optimizer, device):
    model.train()
    total_loss = 0
    correct_predictions = 0
    total_predictions = 0

    progress_bar = tqdm(data_loader, desc="Training")
    for batch in progress_bar:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        optimizer.zero_grad()
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        logits = outputs.logits

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        _, predicted = torch.max(logits, dim=1)
        correct_predictions += (predicted == labels).sum().item()
        total_predictions += labels.size(0)

        # C·∫≠p nh·∫≠t tqdm bar
        accuracy = correct_predictions / total_predictions
        progress_bar.set_postfix(loss=loss.item(), acc=accuracy)

    avg_loss = total_loss / len(data_loader)
    accuracy = correct_predictions / total_predictions
    return avg_loss, accuracy

# Hu·∫•n luy·ªán trong 3 epoch (ho·∫∑c √≠t h∆°n n·∫øu c·∫ßn t·ªëi ∆∞u nhanh)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

epochs = 3
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}/{epochs}")
    train_loss, train_accuracy = train_epoch(model, train_loader, optimizer, device)
    print(f" Train Loss: {train_loss:.4f}, Accuracy: {train_accuracy:.4f}")
```

#L∆∞u l·∫°i gi√° tr·ªã Loss v√† Accuracy




```python
import csv

# Kh·ªüi t·∫°o danh s√°ch ƒë·ªÉ l∆∞u tr·ªØ loss v√† accuracy
train_losses = []
train_accuracies = []

# T·∫°o ho·∫∑c m·ªü file CSV ƒë·ªÉ l∆∞u k·∫øt qu·∫£
with open('/content/drive/MyDrive/training_results.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Epoch', 'Loss', 'Accuracy'])  # Ti√™u ƒë·ªÅ c·ªôt

    # Hu·∫•n luy·ªán m√¥ h√¨nh v√† t√≠nh to√°n loss v√† accuracy
    for epoch in range(3):  # Ch·∫°y qua 3 epoch
        model.train()
        total_loss = 0
        correct_predictions = 0
        total_samples = 0
        for batch in train_loader:
            input_ids = batch['input_ids'].to('cuda')
            attention_mask = batch['attention_mask'].to('cuda')
            labels = batch['labels'].to('cuda')

            optimizer.zero_grad()
            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            total_loss += loss.item()

            # T√≠nh accuracy
            logits = outputs.logits
            preds = torch.argmax(logits, dim=1)
            correct_predictions += (preds == labels).sum().item()
            total_samples += labels.size(0)

            loss.backward()
            optimizer.step()

        avg_loss = total_loss / len(train_loader)
        accuracy = correct_predictions / total_samples
        train_losses.append(avg_loss)
        train_accuracies.append(accuracy)

        # L∆∞u k·∫øt qu·∫£ v√†o file CSV
        writer.writerow([epoch+1, avg_loss, accuracy])

        print(f"Epoch {epoch+1}, Loss: {avg_loss}, Accuracy: {accuracy}")
```

#V·∫Ω ƒë·ªì th·ªã Loss v√† Accuracy

```python
import matplotlib.pyplot as plt

# V·∫Ω ƒë·ªì th·ªã Loss
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  # T·∫°o subplot ƒë·ªÉ v·∫Ω 2 ƒë·ªì th·ªã
plt.plot(range(1, len(train_losses) + 1), train_losses, marker='o', color='red', label='Train Loss')
plt.title('Loss trong qu√° tr√¨nh hu·∫•n luy·ªán')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

# V·∫Ω ƒë·ªì th·ªã Accuracy
plt.subplot(1, 2, 2)
plt.plot(range(1, len(train_accuracies) + 1), train_accuracies, marker='o', color='green', label='Train Accuracy')
plt.title('Accuracy trong qu√° tr√¨nh hu·∫•n luy·ªán')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend()

# Hi·ªÉn th·ªã ƒë·ªì th·ªã
plt.tight_layout()
plt.show()
```

#ƒê√°nh gi√° m√¥ h√¨nh tr√™n t·∫≠p ki·ªÉm tra

```python
from sklearn.metrics import classification_report

model.eval()
predictions = []
true_labels = []

with torch.no_grad():
    for batch in test_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        predictions.append(torch.argmax(logits, dim=1))
        true_labels.append(labels)

# ƒê√°nh gi√°
predictions = torch.cat(predictions).cpu().numpy()
true_labels = torch.cat(true_labels).cpu().numpy()

print(classification_report(true_labels, predictions, target_names=['negative', 'neutral', 'positive']))
```
